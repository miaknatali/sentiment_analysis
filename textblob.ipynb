```python
import pandas as pd
df = pd.read_csv("data/test.csv", encoding = "latin-1")
df = df[["text", "sentiment"]]
df.head()
```

```python
from textblob import TextBlob

def textblob_score(text):
    blob = TextBlob(text)
    return blob.sentiment
df["textblob"]= df["text"].apply(textblob_score)
df["polarity"] = df["textblob"].apply(lambda x: x.polarity)

def textblob_label(polarity):
    if polarity >= 0.05:
        return "positive"
    elif polarity <= -0.05:
        return "negative"
    else:
        return "neutral"
df["label"] = df["polarity"].apply(textblob_label)

TP = FP = TN = FN = T0 = F0 = 0
def compare(row):
    global TP, FP, TN, FN, T0, F0

    trueVal = row["sentiment"]
    textblobVal = row["label"]

    if textblobVal == "positive":
        if textblobVal == trueVal:
            TP += 1
        else:
            FP += 1
    elif textblobVal == "negative":
        if textblobVal == trueVal:
            TN += 1
        else:
            FN += 1
    else:
        if textblobVal == trueVal:
            T0 += 1
        else:
            F0 += 1

df.apply(compare, axis = 1)
accuracy = (TP+TN+T0)/(TP+FP+TN+FN+T0+F0)

print("Accuracy: ", accuracy)
df.head()
```
